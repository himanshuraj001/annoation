```python
import cv2
import numpy as np
```

```python
crop=False
ix=-1
iy=-1
ref=[]
```

```python
def shape(event, x, y, flags, param):
    global crop,ref,ix,iy
    if event == cv2.EVENT_LBUTTONDOWN:
        ix,iy=x,y
        crop = True
        
    elif event == cv2.EVENT_LBUTTONUP :
        crop=False
        cv2.rectangle(img, (ix, iy), (x, y), (0, 0, 0), 2)
        ref.append([ix,iy,x,y])
    cv2.imshow("image", img)   
```

```python
img=cv2.imread("Untitled.png")
clone=img.copy()
```

```python
cv2.namedWindow("image")
cv2.setMouseCallback("image",shape)
while True:
    cv2.imshow("image", img)
    key = cv2.waitKey(1) & 0xFF

  # if the 'c' key is pressed, break from the loop
    if key == ord("c"):
        break
cv2.destroyAllWindows()
```

```python
label=[]
for i in range(len(ref)):
    j=input("Enter the label chosen "+str(i+1)+"  ")
    label.append(j)
```

```python
t=[]

for i in ref:
    crop_img=clone[i[1]:i[3],i[0]:i[2]]
    t.append(crop_img)
    

```

```python
f=t[0]
from PIL import Image
img = Image.fromarray(f, 'RGB')
img
```

```python
import  pickle
output = open('data.pkl', 'wb')
```

```python
for x,y in zip(t,label):
    z={}
    z["img"]=x
    z["label"]=y
    pickle.dump(z, output)

    
output.close()
```
